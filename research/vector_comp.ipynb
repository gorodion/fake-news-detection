```python
from sklearn.feature_extraction.text import TfidfVectorizer as Vect
vect = Vect(max_features=10000)
X = vect.fit_transform(df.text.fillna(''))
```

```python
cosine = metrics.pairwise.cosine_similarity(X)
```

```python
cosine = np.triu(cosine, 1)
```

```python
print(df.iloc[0].text)
```

```python
cosine.argmax(1)
```
